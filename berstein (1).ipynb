```python
from qiskit import *
%matplotlib inline

secret = '101001'
```

```python
circuit = QuantumCircuit(6+1, 6)
circuit.draw(output = 'mpl')
```

```python
for i in range(6):
    circuit.h(i)
circuit.x(6)
circuit.h(6)

circuit.barrier()

circuit.cx(5,6)
circuit.cx(2,6)
circuit.cx(0,6)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
for i in range(6):
    circuit.h(i)
    
circuit.barrier()
for i in range(6):
    circuit.measure(i,i)
    
circuit.draw(output = 'mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1).result()
count = result.get_counts()
print(count)
```
